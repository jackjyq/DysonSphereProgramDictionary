```python
import json
```

```python
with open("./source.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
data["items"]
```

```python
items = {v for v in data["items"].values()}
items
```

```python
recipes = {v for v in data["recipes"].values()}
recipes
```

```python
def is_chinese(char: str) -> bool:
    """检查字符是汉字

    参考: https://cloud.tencent.com/developer/article/1499958
    """
    assert len(char) == 1
    return "\u4e00" <= char <= "\u9fa5"
```

```python
lexicon = items | recipes
lexicon
```

```python
transformed_lexicon = set()
for word in lexicon:
    transformed_word = ""
    for char in word:
        if is_chinese(char):
            transformed_word += char
        else:
            break
    if len(transformed_word) > 1:
        transformed_lexicon.add(transformed_word)

transformed_lexicon
```

```python
with open("./dictionary.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(transformed_lexicon))
```
