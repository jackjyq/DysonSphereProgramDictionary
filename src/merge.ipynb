```python
import glob

dictionary = set()
for file in glob.glob("./**/dictionary*.txt", recursive=True):
    with open(file, "r", encoding="utf-8") as f:
        lines = f.readlines()
        print(f"{file} has {len(lines)} words")
        for line in lines:
            dictionary.add(line.strip())
        print(f"Dictionary has {len(dictionary)} words")

dictionary
```

```python
len(dictionary)
```

```python
dictionary = sorted(dictionary)
```

```python
with open("dictionary.txt", "w", encoding="utf-8") as f:
    for word in dictionary:
        f.write(f"{word}\n")
```
